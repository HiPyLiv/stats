```python
def pie_calc(data, key):
    '''
    Analyse data and return labels and frequency
    '''
    labels = []
    for item in data:
        labels.append(item[key])
    
    labels_set = set(labels)
    
    freq_list = []
    
    for l in labels_set:
        count = 0
        for item in data:
            if l == item[key]:
                count+=1
        freq_list.append(count)
    
    return labels_set, freq_list
```
